# **Read dataset**

```
def read_dataset(fname):
    sentences = []
    tags = []
    with open(fname) as f:
        content = f.readlines()
    # you may also want to remove whitespace characters like `\n` at the end of each line
    content = [x.strip() for x in content]
    idx_line = 2 # skip keterangan #Format dan baris kosong setelahnya
    eof_stat = False
    while idx_line < len(content):
        sent = []
        tag = []
        print('idx_line =')
        print(idx_line)

        while len(content[idx_line]) > 0 and not eof_stat:
            print("content = ")
            print(content[idx_line])
            if  not content[idx_line].startswith('#Text='):
                content_part = content[idx_line].split('\t')
                sent.append(content_part[2].lower())
                tag.append(content_part[3])
            idx_line = idx_line + 1
            if (idx_line == len(content)-1):
                eof_stat= True
        sentences.append(sent)
        tags.append(tag)
        idx_line = idx_line+2        
    return sentences, tags
```

```
sentences,tags = read_dataset('annotated_corpus_POSTag.tsv')
```

```
def wordTag(sentences, tags):
  wordTag = {}
  for i in range(len(sentences)):
    for j in range (len(sentences[i])):
      join = str(sentences[i][j]+','+tags[i][j])
      if join in wordTag.keys():
        wordTag[join] = wordTag[join]+1
      else:
        wordTag[join] = 1
  return wordTag
```

```
word_tag = wordTag(sentences, tags)
word_tag
```

```
def countTag(tags, wordTag):
  tag_count = {}
  val = 0
  for i in range(len(tags)):
    for j in range(len(tags[i])):
      if tags[i][j] in tag_count.keys():
        tag_count[tags[i][j]] = tag_count[tags[i][j]] + 1
      else:
        tag_count[tags[i][j]] = 1
  tag_count['<start>'] = len(sentences)
  return tag_count
```

```
tag_count = countTag(tags, word_tag)
tag_count
```

```
def tagTrans(sentences, tags):
  trans_tag = {}
  for i in range(len(sentences)):
    for j in range(len(tags[i])):
      if j == 0:
        current_tag = '<start>,' + tags[i][j]
      else:
        current_tag = prev_tag + ',' + tags[i][j]
      prev_tag = tags[i][j]
      if current_tag in trans_tag.keys():
        trans_tag[current_tag] = trans_tag[current_tag] + 1
      else:
        trans_tag[current_tag] = 1
  return trans_tag
```

```
trans_tag = tagTrans(sentences,tags)
trans_tag
```

# **Viterbi Method**

```
def create_trans_prob_table(trans_tag, tag_count):
    print(trans_tag)
    trans_prob = {}
    for tag1 in tag_count.keys():
        for tag2 in tag_count.keys():
            trans_idx = tag1+','+tag2
            if trans_idx in trans_tag:
                trans_prob[trans_idx] = trans_tag[trans_idx]/tag_count[tag1]
    return trans_prob
```

```
trans_prob = create_trans_prob_table(trans_tag, tag_count)
trans_prob
```

```
def create_emission_prob_table(word_tag, tag_count):
    emission_prob = {}
    for word_tag_entry in word_tag.keys():
        word_tag_split = re.split(r'[,](?=[a-zA-Z|\*])', word_tag_entry)
        current_word = word_tag_split[0]
        current_tag = word_tag_split[1]
        emission_key = current_word+','+current_tag
        emission_prob[emission_key] = word_tag[word_tag_entry]/tag_count[current_tag]     
    return emission_prob
```

```
emission_prob = create_emission_prob_table(word_tag, tag_count)
emission_prob
```

```
def viterbi(trans_prob, emission_prob, tag_count, sentence):
  tag_sequence = []
  words = nltk.word_tokenize(sentence.lower())
  tag_count_new = tag_count.copy()
  tag_count_new.pop('<start>')
  tag_count_new.pop('*')
  maxTag = ''
  viterbi_mat_prev = {}
  for i in range(len(words)):
    viterbi_mat = {}
    maxVal = 0
    for j in tag_count_new:
      if (words[i]+','+j) in emission_prob.keys():
        viterbi_mat[j] = emission_prob[words[i]+','+j]
    if viterbi_mat_prev == {}:
      for k in viterbi_mat.keys():
        if trans_prob['<start>,'+k] > maxVal:
          maxVal = trans_prob['<start>,'+k]
          maxTag = k
      viterbi_mat['prevTag'] = '<start>'
      viterbi_mat_prev = viterbi_mat.copy()
      viterbi_mat[maxTag] = maxVal
      tag_sequence.append(viterbi_mat)
    else:
      viterbi_mat_new = {}
      for l in viterbi_mat.keys():
        viterbi_mat_new['prevTag,'+l] = ''
        for m in viterbi_mat_prev.keys():
          if (m+','+l) in trans_prob.keys():
            if trans_prob[m+','+l] > maxVal:
              maxVal = trans_prob[m+','+l]
              prevTag = m
            viterbi_mat_new[l] = maxVal*viterbi_mat[l]*trans_prob[m+','+l]
        viterbi_mat_new['prevTag,'+l] = prevTag
      if viterbi_mat_new == {}:
        viterbi_mat_new['NNO'] = 1
        maxVal = 0
        for p in viterbi_mat_prev.keys():
          if (p+',NNO') in trans_prob.keys():
            if trans_prob[p+',NNO'] > maxVal:
              maxVal = trans_prob[p+',NNO']
              prevTag = p
            viterbi_mat_new['NNO'] = maxVal*viterbi_mat_new['NNO']*trans_prob[p+',NNO']
        viterbi_mat_new['prevTag,NNO'] = prevTag
      viterbi_mat_prev = viterbi_mat_new.copy()
      tag_sequence.append(viterbi_mat_new)
  result = []
  finalMax = 0
  finalTag = ''
  for z in tag_sequence[len(tag_sequence)-1].keys():
    if ',' not in z:
      if tag_sequence[len(tag_sequence)-1][z] > finalMax:
        finalMax = tag_sequence[len(tag_sequence)-1][z]
        finalTag = z

  j = len(tag_sequence)-1
  while j >= 0:
    res = finalTag
    result.insert(0,res)
    if j != 0:
      finalTag = tag_sequence[j]['prevTag,'+res]
    else:
      finalTag = tag_sequence[j]['prevTag']
    j = j - 1
  result.insert(0,'<start>')
  return result
```

```
viterbi(trans_prob, emission_prob, tag_count,'Halaman ini memuat daftar benda cagar budaya di Malaysia.')
```

# **Classification-based Method**

```
def features(sentence, index):
    """ sentence: [w1, w2, ...], index: the index of the word """
    # print("sentence index = ")
    # print(sentence[index])
    return {
        'word': sentence[index],
        # 'prefix-1': sentence[index][0],
        # 'prefix-2': sentence[index][:2],
        # 'prefix-3': sentence[index][:3],
        # 'suffix-1': sentence[index][-1],
        # 'suffix-2': sentence[index][-2:],
        # 'suffix-3': sentence[index][-3:],
        'prev_word': '' if index == 0 else sentence[index - 1],
        'next_word': '' if index == len(sentence) - 1 else sentence[index + 1],
    }


 
def transform_to_dataset(sentences, tags):
    X, y = [], []
 
    for sentence_idx in range(len(sentences)):
        for index in range(len(sentences[sentence_idx])):
            X.append(features(sentences[sentence_idx], index))
            y.append(tags[sentence_idx][index])
 
    return X, y
 

# tes baca file 
sentences,tags = read_dataset('annotated_corpus_POSTag.tsv')
# print(sentences[0])
# print(tags[0])
# print(len(sentences[0]))
# print(len(tags[0]))

# Split the dataset for training and testing
cutoff = int(.75 * len(sentences))
training_sentences = sentences[:cutoff]
test_sentences = sentences[cutoff:]
training_tags = tags[:cutoff]
test_tags = tags[cutoff:]
 
#print(len(training_sentences))   
#print(len(test_sentences))         

X, y = transform_to_dataset(training_sentences, training_tags)
print('data training ke-0 =')
print(X[0])
print('label training ke-0 =')
print(y[0])

from sklearn import tree
from sklearn.feature_extraction import DictVectorizer
from sklearn.pipeline import Pipeline
 
clf = Pipeline([
    ('vectorizer', DictVectorizer(sparse=False)),
    ('classifier', tree.DecisionTreeClassifier(criterion='entropy'))
])
clf.fit(X, y)   
 
print('Training completed')

X_test, y_test = transform_to_dataset(test_sentences, test_tags)
 
print("Accuracy:")
print(clf.score(X_test, y_test))

from nltk import word_tokenize

def pos_tag(sentence):
    tags = clf.predict([features(sentence, index) for index in range(len(sentence))])
    return tags
 
print(pos_tag(word_tokenize('Halaman ini memuat daftar benda cagar budaya di Malaysia.')))
```
